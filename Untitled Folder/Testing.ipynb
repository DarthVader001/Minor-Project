```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
import numpy as np
from math import sqrt
df_clean=pd.read_csv('Uber_Customer_Support_Tickets_Prepared.csv')
print(df_clean.head())
print(df_clean.columns)
```

```python
# Treatment = Chatbot, Control = LiveAgent
chatbot = df_clean[df_clean['Channel'].str.lower() == 'chatbot']
liveagent = df_clean[df_clean['Channel'].str.lower() == 'liveagent']

print("Chatbot tickets:", len(chatbot))
print("LiveAgent tickets:", len(liveagent))
```

```python
print("Chatbot Mean CSAT:", chatbot['CSAT_Score'].mean())
print("LiveAgent Mean CSAT:", liveagent['CSAT_Score'].mean())

print("Chatbot Mean Resolution Time:", chatbot['Resolution_Time_Minutes'].mean())
print("LiveAgent Mean Resolution Time:", liveagent['Resolution_Time_Minutes'].mean())
```

```python
# CSAT comparison
t_cs, p_cs = stats.ttest_ind(chatbot['CSAT_Score'], liveagent['CSAT_Score'], equal_var=False)
print(f"CSAT t-test → t={t_cs:.3f}, p={p_cs:.3f}")

# Resolution Time comparison
t_rt, p_rt = stats.ttest_ind(chatbot['Resolution_Time_Minutes'], liveagent['Resolution_Time_Minutes'], equal_var=False)
print(f"Resolution Time t-test → t={t_rt:.3f}, p={p_rt:.3f}")
```

```python
# Create binary outcome: Low CSAT = 1 if <=3, else 0
df_clean['Low_CSAT'] = (df_clean['CSAT_Score'] <= 3).astype(int)

contingency = pd.crosstab(df_clean['Channel'], df_clean['Low_CSAT'])
print(contingency)

chi2, p_chi, dof, expected = stats.chi2_contingency(contingency)
print(f"Chi-square test for Low CSAT proportions → chi2={chi2:.3f}, p={p_chi:.3f}")
```

```python
# Average CSAT bar chart
avg_scores = df_clean.groupby("Channel")["CSAT_Score"].mean()

avg_scores.plot(kind="bar", color=["skyblue","orange"], edgecolor="black")
plt.title("Average CSAT by Channel")
plt.ylabel("Mean CSAT Score")
plt.show()

# Boxplot for resolution time
df_clean.boxplot(column="Resolution_Time_Minutes", by="Channel")
plt.title("Resolution Time by Channel")
plt.suptitle("")  # removes default subtitle
plt.ylabel("Minutes")
plt.show()
```
