```python
from sklearn.ensemble import IsolationForest
import pandas as pd
```

```python
df = pd.read_csv("Uber_Customer_Support_Tickets_Prepared.csv")
```

```python
df = pd.read_csv("Uber_Customer_Support_Tickets_Prepared.csv")
```

```python
# Isolation Forest
X = df[['Response_Time_Minutes','Resolution_Time_Minutes']]
iso = IsolationForest(contamination=0.05, random_state=42)
df['Anomaly_Flag'] = iso.fit_predict(X)
df['Anomaly_Flag'] = df['Anomaly_Flag'].map({1:0, -1:1})  # 1 = anomaly
```

```python
# Show anomalies
anomalies = df[df['Anomaly_Flag']==1]
print(f"Detected {len(anomalies)} anomalies")
anomalies[['Ticket_ID','Response_Time_Minutes','Resolution_Time_Minutes','Anomaly_Flag']]
```
